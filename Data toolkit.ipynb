NumPy Array Functions

np.array:

This function creates a new NumPy array from a given input (like lists, tuples, etc.).
It copies the data into a new array, meaning modifications to the original data won't affect the new array.

np.asarray:

This function converts the input to an array, but it does not create a new copy if the input is already an array (i.e., if the input is a NumPy array, it will return it as-is).
Itâ€™s generally more efficient than np.array because it avoids unnecessary copying.

np.asanyarray:

Similar to np.asarray, but it allows for subclasses of arrays (like masked arrays) to remain as their respective types.
If the input is a subclass of ndarray, it will return that subclass instead of a base ndarra.

Copying in Python
Shallow Copy:

A shallow copy creates a new object but does not create copies of the nested objects. Instead, it copies references to those nested objects.

Changes to mutable nested objects in the original will reflect in the shallow copy (and vice versa).

Deep Copy:

A deep copy creates a new object and recursively copies all nested objects, creating entirely independent copies.

Changes to the original object or its nested objects do not affect the deep copy.

```python
import numpy as np

# Create a 5x6 array with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(5, 6))

# Display the original array
print("Original Array:")
print(random_array)

# a) Extract all even integers from the array
even_integers = random_array[random_array % 2 == 0]
print("\nEven Integers:")
print(even_integers)

# b) Extract all odd integers from the array
odd_integers = random_array[random_array % 2 != 0]
print("\nOdd Integers:")
print(odd_integers)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
dates = pd.date_range(start='2023-01-01', periods=100)

# Create random data for Temperature and Humidity
temperature = np.random.uniform(low=15, high=35, size=len(dates))  # Temperature between 15 and 35 degrees
humidity = np.random.uniform(low=30, high=90, size=len(dates))    # Humidity between 30% and 90%

# Create a DataFrame
data = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Set the Date as the index
data.set_index('Date', inplace=True)

# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(data.index, data['Temperature'], color='tab:red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(data.index, data['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# b) Label the x-axis
ax1.set_xlabel('Date')

# c) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay a line plot representing the normal distribution's PDF
xmin, xmax = plt.xlim()  # Get the limits of the x-axis
x = np.linspace(xmin, xmax, 100)  # Generate values for x-axis
p = norm.pdf(x, loc=0, scale=1)  # Calculate the PDF for the normal distribution
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# c) Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)

# Create a DataFrame to hold the data and add a 'Quadrant' column
data = pd.DataFrame({'X': x, 'Y': y})

# Define the quadrant based on the position of (X, Y)
def determine_quadrant(row):
    if row['X'] > 0 and row['Y'] > 0:
        return 'Quadrant 1'
    elif row['X'] < 0 and row['Y'] > 0:
        return 'Quadrant 2'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant 3'
    elif row['X'] > 0 and row['Y'] < 0:
        return 'Quadrant 4'
    else:
        return 'Origin'

data['Quadrant'] = data.apply(determine_quadrant, axis=1)

# Create a Seaborn scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='Set1', style='Quadrant', markers=['o', 's', 'D', '^'], s=100)

# Add legend, labels, and title
plt.legend(title='Quadrants')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display the plot
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output for Jupyter Notebook (remove if running in a script)
output_notebook()

# Generate x values
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2Ï€
y = np.sin(x)  # Calculate sine of x

# Create a new plot
p = figure(title='Sine Wave Function', x_axis_label='X-axis (radians)', y_axis_label='Y-axis (amplitude)',
           width=800, height=400)

# Add grid lines
p.grid.grid_line_alpha = 0.3  # Set the transparency of grid lines

# Plot the sine wave
p.line(x, y, line_width=2, color='blue', legend_label='Sine Wave')

# Customize legend
p.legend.location = "top_left"
p.legend.click_policy = "hide"

# Show the plot
show(p)
```
